# Data Description

The dataset used for this visualization assignment contains *automobile_sales* data representing automobile sales and related variables during recession and non-recession period. 

The dataset includes the following variables:
<br>1. Date: The month end date of sales observation.
<br>2. Recession: A binary variable indicating recession perion; 1 means it was recession, 0 means it was normal.
<br>3. Automobile_Sales: The number of vehicles sold during the period.
<br>4. GDP: The per capita GDP value in USD.
<br>5. Unemployment_Rate: The monthly unemployment rate.
<br>6. Consumer_Confidence: A synthetic index representing consumer confidence, which can impact consumer spending and automobile purchases.
<br>7. Seasonality_Weight: The weight representing the seasonality effect on automobile sales during the period. This variable represents the seasonal effect on automobile sales for a given month. In the automobile industry, sales often fluctuate throughout the year due to seasonal patterns—for example, sales may increase during the festive season or year-end promotions, and decrease during off-peak months like post-holiday winter periods.
A value greater than 1 indicates higher-than-average sales expected for that month due to seasonal trends (e.g., holiday season or new model launches).
A value less than 1 suggests lower-than-average sales due to seasonal slowdowns.
A value around 1 means the season has neutral or average effect on sales.
<br>8. Price: The average vehicle price during the period.
<br>9. Advertising_Expenditure: The advertising expenditure of the company.
<br>10.Vehicle_Type: The type of vehicles sold; Supperminicar, Smallfamiliycar,                 Mediumfamilycar, Executivecar, Sports.
<br>11.Competition: The measure of competition in the market, such as the number of competitors or market share of major manufacturers.
<br>12.Month: Month of the observation extracted from Date..
<br>13.Year: Year of the observation extracted from Date.
<br>
By examining various factors mentioned above from the dataset, you aim to gain insights into how recessions impacted automobile sales for your company.

````python
# Task 1.2: Advertising Expenditure vs Automobile Sales (Non-Recession Periods)

# Filter for non-recession periods
df_non_recession = df[df['Recession'] == 0]

# Plot
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot Automobile Sales on primary y-axis
color1 = 'blue'
ax1.set_xlabel('Year')
ax1.set_ylabel('Automobile Sales', color=color1)
ax1.plot(df_non_recession.groupby('Year')['Automobile_Sales'].mean(), 
         color=color1, marker='o', label='Automobile Sales')
ax1.tick_params(axis='y', labelcolor=color1)

# Plot Advertising Expenditure on secondary y-axis
ax2 = ax1.twinx()
color2 = 'orange'
ax2.set_ylabel('Advertising Expenditure', color=color2)
ax2.plot(df_non_recession.groupby('Year')['Advertising_Expenditure'].mean(), 
         color=color2, marker='s', linestyle='--', label='Advertising Expenditure')
ax2.tick_params(axis='y', labelcolor=color2)

# Title and layout
plt.title('Advertising Expenditure vs Automobile Sales\nDuring Non-Recession Periods')
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.9))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

### TASK 1.3: Use the functionality of **Seaborn Library** to create a visualization to compare the sales trend per vehicle type for a recession period with a non-recession period.


<details><summary>Click here for a hint</summary>

<p>
     To visualize the average number of vehicles sold during recession and non-recession periods, you can use a bar chart
        <br> You will need to group recession average Automobile_Sales and then plot it<br>

````python
# Task 1.3: Sales Trend per Vehicle Type - Recession vs Non-Recession (Seaborn)

# Create a new column to label recession periods
df['Recession_Period'] = df['Recession'].map({1: 'Recession', 0: 'Non-Recession'})

# Calculate average sales by Year, Vehicle Type and Recession Period
df_grouped = df.groupby(['Year', 'Vehicle_Type', 'Recession_Period'])['Automobile_Sales'].mean().reset_index()

# Plot using Seaborn FacetGrid
g = sns.FacetGrid(df_grouped, 
                  col='Recession_Period', 
                  hue='Vehicle_Type',
                  height=6, 
                  aspect=1.2,
                  palette='tab10')

g.map(sns.lineplot, 'Year', 'Automobile_Sales', marker='o')

# Titles and labels
g.set_axis_labels('Year', 'Average Automobile Sales')
g.set_titles(col_template='{col_name}')
g.add_legend(title='Vehicle Type', bbox_to_anchor=(1.05, 0.5))
g.fig.suptitle('Sales Trend per Vehicle Type:\nRecession vs Non-Recession Period', 
               y=1.02, fontsize=14, fontweight='bold')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

### TASK 1.4: Use sub plotting to compare the variations in GDP during recession and non-recession period by developing line plots for each period.
<br>Now, you want to find more insights from the data to understand the reason. <br>Plot a two line charts using subplotting to answer:-
#### How did the GDP vary over time during recession and non-recession periods? 
<br>Make use of <u>add_subplot()</u> from Matplotlib for this comparision.


````python
# Task 1.4: GDP Variation - Recession vs Non-Recession using add_subplot()

# Filter data for recession and non-recession periods
df_recession = df[df['Recession'] == 1]
df_non_recession = df[df['Recession'] == 0]

# Calculate average GDP per year for each period
gdp_recession = df_recession.groupby('Year')['GDP'].mean().reset_index()
gdp_non_recession = df_non_recession.groupby('Year')['GDP'].mean().reset_index()

# Create figure
fig = plt.figure(figsize=(14, 8))

# ── Subplot 1: Recession Period ───────────────────────────────────────────────
ax1 = fig.add_subplot(1, 2, 1)
ax1.plot(gdp_recession['Year'], 
         gdp_recession['GDP'],
         color='red',
         marker='o',
         linewidth=2,
         markersize=6,
         label='GDP (Recession)')

ax1.set_title('GDP Variation During\nRecession Period', 
              fontsize=13, fontweight='bold')
ax1.set_xlabel('Year', fontsize=11)
ax1.set_ylabel('Average GDP', fontsize=11)
ax1.legend(loc='upper left')
ax1.grid(True, linestyle='--', alpha=0.6)
ax1.tick_params(axis='x', rotation=45)

# Annotate min and max points
min_idx = gdp_recession['GDP'].idxmin()
max_idx = gdp_recession['GDP'].idxmax()
ax1.annotate(f"Min: {gdp_recession['GDP'][min_idx]:.2f}",
             xy=(gdp_recession['Year'][min_idx], gdp_recession['GDP'][min_idx]),
             xytext=(10, 10), textcoords='offset points', color='darkred', fontsize=9)
ax1.annotate(f"Max: {gdp_recession['GDP'][max_idx]:.2f}",
             xy=(gdp_recession['Year'][max_idx], gdp_recession['GDP'][max_idx]),
             xytext=(10, -15), textcoords='offset points', color='darkred', fontsize=9)

# ── Subplot 2: Non-Recession Period ──────────────────────────────────────────
ax2 = fig.add_subplot(1, 2, 2)
ax2.plot(gdp_non_recession['Year'], 
         gdp_non_recession['GDP'],
         color='green',
         marker='s',
         linewidth=2,
         markersize=6,
         label='GDP (Non-Recession)')

ax2.set_title('GDP Variation During\nNon-Recession Period', 
              fontsize=13, fontweight='bold')
ax2.set_xlabel('Year', fontsize=11)
ax2.set_ylabel('Average GDP', fontsize=11)
ax2.legend(loc='upper left')
ax2.grid(True, linestyle='--', alpha=0.6)
ax2.tick_params(axis='x', rotation=45)

# Annotate min and max points
min_idx2 = gdp_non_recession['GDP'].idxmin()
max_idx2 = gdp_non_recession['GDP'].idxmax()
ax2.annotate(f"Min: {gdp_non_recession['GDP'][min_idx2]:.2f}",
             xy=(gdp_non_recession['Year'][min_idx2], gdp_non_recession['GDP'][min_idx2]),
             xytext=(10, 10), textcoords='offset points', color='darkgreen', fontsize=9)
ax2.annotate(f"Max: {gdp_non_recession['GDP'][max_idx2]:.2f}",
             xy=(gdp_non_recession['Year'][max_idx2], gdp_non_recession['GDP'][max_idx2]),
             xytext=(10, -15), textcoords='offset points', color='darkgreen', fontsize=9)

# ── Overall Figure Title ──────────────────────────────────────────────────────
fig.suptitle('GDP Variation Over Time:\nRecession vs Non-Recession Period', 
             fontsize=15, fontweight='bold', y=1.02)

plt.tight_layout()
plt.show()
````

### TASK 1.5: Develop a Bubble plot for displaying the impact of seasonality on Automobile Sales.
<br>How has seasonality impacted the sales, in which months the sales were high or low? Check it for non-recession years to understand the trend

##### Develop a Bubble plot for displaying Automobile Sales for every month and use Seasonality Weight for representing the size of each bubble<br>
Title this plot as 'Seasonality impact on Automobile Sales'


<details><summary>Click here for a hint</summary>

<p>
     You can create Bubble Chart by calling the scatter()
        <br>Pass the 'Month' and 'Automobile_Sales' to the functions as x and y and then use Seasonality weight for size parameter</p>
</details>


````python
# Task 1.5: Bubble Plot - Seasonality Impact on Automobile Sales (Non-Recession)

# Filter for non-recession periods
df_non_recession = df[df['Recession'] == 0]

# Calculate average sales and seasonality weight per month
df_bubble = df_non_recession.groupby('Month').agg(
    Automobile_Sales=('Automobile_Sales', 'mean'),
    Seasonality_Weight=('Seasonality_Weight', 'mean')
).reset_index()

# Define month order
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
               'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Map months to numeric for proper ordering on x-axis
df_bubble['Month_Num'] = df_bubble['Month'].map(
    {m: i+1 for i, m in enumerate(month_order)}
)

# Sort by month number
df_bubble = df_bubble.sort_values('Month_Num')

# ── Plot ──────────────────────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(14, 7))

# Create bubble plot using scatter
bubble = ax.scatter(
    df_bubble['Month_Num'],
    df_bubble['Automobile_Sales'],
    s=df_bubble['Seasonality_Weight'] * 1000,  # scale bubble size
    c=df_bubble['Automobile_Sales'],            # color by sales value
    cmap='coolwarm',
    alpha=0.7,
    edgecolors='black',
    linewidth=1.2
)

# ── Colorbar ──────────────────────────────────────────────────────────────────
cbar = plt.colorbar(bubble, ax=ax)
cbar.set_label('Automobile Sales', fontsize=11)

# ── Annotate each bubble with month name ──────────────────────────────────────
for _, row in df_bubble.iterrows():
    ax.annotate(
        month_order[int(row['Month_Num']) - 1],
        xy=(row['Month_Num'], row['Automobile_Sales']),
        ha='center',
        va='center',
        fontsize=9,
        fontweight='bold',
        color='white'
    )

# ── Axes and Labels ───────────────────────────────────────────────────────────
ax.set_xticks(range(1, 13))
ax.set_xticklabels(month_order, fontsize=11)
ax.set_xlabel('Month', fontsize=12)
ax.set_ylabel('Average Automobile Sales', fontsize=12)
ax.set_title('Seasonality impact on Automobile Sales',
             fontsize=15, fontweight='bold')
ax.grid(True, linestyle='--', alpha=0.5)

# ── Legend for bubble size ────────────────────────────────────────────────────
for size_val, label in zip([0.5, 1.0, 1.5], ['Low', 'Medium', 'High']):
    ax.scatter([], [], 
               s=size_val * 1000, 
               c='gray', 
               alpha=0.5, 
               label=f'Seasonality: {label}')

ax.legend(title='Seasonality Weight',
          loc='upper left',
          fontsize=10,
          title_fontsize=11)

plt.tight_layout()
plt.show()
````

 ### TASK 1.6: Use the functionality of Matplotlib to develop a scatter plot to identify the relationship between consumer confidence and automobile sales during recessions.
 #### From the data, develop a scatter plot to identify the relationship between consumer confidence and automobile sales during recession periods.
 <br> Title this plot as 'Consumer Confidence and Automobile Sales during Recessions'


<details><summary>Click here for a hint</summary>

<p>
     You can create dataframe where recession is '1'.
        <br>Pass the 'Consumer_Confidence' and 'Automobile_Sales' to the plt.scatter()</p>

````python
# Task 1.6: Scatter Plot - Consumer Confidence vs Automobile Sales (Recession)

# Filter for recession periods only
df_recession = df[df['Recession'] == 1]

# ── Plot ──────────────────────────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(12, 7))

# Create scatter plot
scatter = ax.scatter(
    df_recession['Consumer_Confidence'],
    df_recession['Automobile_Sales'],
    c=df_recession['Automobile_Sales'],   # color by sales value
    cmap='RdYlGn',                         # red=low, green=high
    s=100,                                 # bubble size
    alpha=0.7,
    edgecolors='black',
    linewidth=0.8
)

# ── Colorbar ──────────────────────────────────────────────────────────────────
cbar = plt.colorbar(scatter, ax=ax)
cbar.set_label('Automobile Sales', fontsize=11)

# ── Trend Line (Linear Regression) ───────────────────────────────────────────
import numpy as np

x = df_recession['Consumer_Confidence']
y = df_recession['Automobile_Sales']

# Calculate trend line
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
x_line = np.linspace(x.min(), x.max(), 100)

ax.plot(x_line, p(x_line),
        color='blue',
        linewidth=2,
        linestyle='--',
        label=f'Trend Line (slope: {z[0]:.2f})')

# ── Annotations for Outliers ──────────────────────────────────────────────────
# Annotate top 3 highest sales points
top3 = df_recession.nlargest(3, 'Automobile_Sales')
for _, row in top3.iterrows():
    ax.annotate(
        f"Sales: {row['Automobile_Sales']:.0f}",
        xy=(row['Consumer_Confidence'], row['Automobile_Sales']),
        xytext=(10, 10),
        textcoords='offset points',
        fontsize=8,
        color='darkgreen',
        arrowprops=dict(arrowstyle='->', color='darkgreen', lw=1)
    )

# Annotate top 3 lowest sales points
bottom3 = df_recession.nsmallest(3, 'Automobile_Sales')
for _, row in bottom3.iterrows():
    ax.annotate(
        f"Sales: {row['Automobile_Sales']:.0f}",
        xy=(row['Consumer_Confidence'], row['Automobile_Sales']),
        xytext=(10, -15),
        textcoords='offset points',
        fontsize=8,
        color='darkred',
        arrowprops=dict(arrowstyle='->', color='darkred', lw=1)
    )

# ── Axes and Labels ───────────────────────────────────────────────────────────
ax.set_xlabel('Consumer Confidence', fontsize=12)
ax.set_ylabel('Automobile Sales', fontsize=12)
ax.set_title('Consumer Confidence and Automobile Sales during Recessions',
             fontsize=14, fontweight='bold')
ax.legend(fontsize=10)
ax.grid(True, linestyle='--', alpha=0.5)

# ── Correlation Coefficient ───────────────────────────────────────────────────
correlation = df_recession['Consumer_Confidence'].corr(df_recession['Automobile_Sales'])
ax.text(0.05, 0.95,
        f'Correlation: {correlation:.2f}',
        transform=ax.transAxes,
        fontsize=11,
        verticalalignment='top',
        bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.8))

plt.tight_layout()
plt.show()

# Print correlation summary
print(f"\nCorrelation between Consumer Confidence and Automobile Sales: {correlation:.2f}")
if correlation > 0.5:
    print("Strong Positive Correlation: Higher consumer confidence leads to more automobile sales")
elif correlation > 0:
    print("Weak Positive Correlation: Slight tendency for sales to rise with confidence")
elif correlation < -0.5:
    print("Strong Negative Correlation: Lower consumer confidence leads to fewer automobile sales")
else:
    print("Weak Negative Correlation: Slight tendency for sales to drop with lower confidence")
````

 ### TASK 1.7: Create a pie chart to display the portion of advertising expenditure of XYZAutomotives during recession and non-recession periods.
 <br>How did the advertising expenditure of XYZAutomotives change during recession and non-recession periods? 


<details><summary>Click here for a hint</summary>

<p>
     You can create two dataframe for recession and nonreccession period.
    <br> Calculate the sum of Advertising_Expenditure for both dataframes
    <br> Pass these total values to plt.pie(). May include labels as ['Recession', 'Non-Recession']
        <br>Feel Free to customie the pie further
    <br>title this plot as  - Advertising Expenditure during Recession and Non-Recession Periods</p>
</details>


````python
# Task 1.7: Pie Chart - Advertising Expenditure during Recession vs Non-Recession

# Calculate total advertising expenditure for each period
recession_exp = df[df['Recession'] == 1]['Advertising_Expenditure'].sum()
non_recession_exp = df[df['Recession'] == 0]['Advertising_Expenditure'].sum()

# Data for pie chart
sizes = [recession_exp, non_recession_exp]
labels = ['Recession', 'Non-Recession']
colors = ['#ff6b6b', '#51cf66']       # red for recession, green for non-recession
explode = (0.1, 0)                     # explode recession slice for emphasis

# ── Plot ──────────────────────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(16, 7))

# ── Pie Chart 1: Overall Split ────────────────────────────────────────────────
wedges, texts, autotexts = axes[0].pie(
    sizes,
    labels=labels,
    colors=colors,
    explode=explode,
    autopct='%1.1f%%',
    shadow=True,
    startangle=140,
    textprops={'fontsize': 12},
    wedgeprops={'edgecolor': 'white', 'linewidth': 2}
)

# Style percentage text
for autotext in autotexts:
    autotext.set_fontsize(13)
    autotext.set_fontweight('bold')
    autotext.set_color('white')

axes[0].set_title('Advertising Expenditure:\nRecession vs Non-Recession',
                  fontsize=13, fontweight='bold', pad=20)

# Add value labels below the chart
axes[0].text(-1.3, -1.4,
             f'Recession Total:     ${recession_exp:,.0f}\n'
             f'Non-Recession Total: ${non_recession_exp:,.0f}\n'
             f'Grand Total:         ${recession_exp + non_recession_exp:,.0f}',
             fontsize=10,
             bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.8))

# ── Pie Chart 2: Yearly Breakdown ─────────────────────────────────────────────
df_yearly = df.groupby(['Year', 'Recession'])['Advertising_Expenditure'].sum().reset_index()
df_yearly['Period'] = df_yearly['Recession'].map({1: 'Recession', 0: 'Non-Recession'})

# Get unique years and their expenditure
yearly_exp = df.groupby('Year')['Advertising_Expenditure'].sum()
year_colors = ['#ff6b6b' if df[df['Year'] == y]['Recession'].mean() > 0.5 
               else '#51cf66' for y in yearly_exp.index]

wedges2, texts2, autotexts2 = axes[1].pie(
    yearly_exp.values,
    labels=yearly_exp.index,
    colors=year_colors,
    autopct='%1.1f%%',
    startangle=140,
    textprops={'fontsize': 9},
    wedgeprops={'edgecolor': 'white', 'linewidth': 1.5},
    pctdistance=0.85
)

for autotext in autotexts2:
    autotext.set_fontsize(8)
    autotext.set_fontweight('bold')

axes[1].set_title('Advertising Expenditure\nBreakdown by Year',
                  fontsize=13, fontweight='bold', pad=20)

# Add legend for year chart
from matplotlib.patches import Patch
legend_elements = [
    Patch(facecolor='#ff6b6b', edgecolor='white', label='Recession Year'),
    Patch(facecolor='#51cf66', edgecolor='white', label='Non-Recession Year')
]
axes[1].legend(handles=legend_elements,
               loc='lower center',
               bbox_to_anchor=(0.5, -0.15),
               ncol=2,
               fontsize=10)

# ── Overall Title ─────────────────────────────────────────────────────────────
fig.suptitle('XYZAutomotives Advertising Expenditure Analysis\nRecession vs Non-Recession Periods',
             fontsize=15, fontweight='bold', y=1.02)

plt.tight_layout()
plt.show()

# ── Print Summary ─────────────────────────────────────────────────────────────
print("=" * 55)
print("   Advertising Expenditure Summary - XYZAutomotives")
print("=" * 55)
print(f"  Recession Period Total:     ${recession_exp:>12,.0f}")
print(f"  Non-Recession Period Total: ${non_recession_exp:>12,.0f}")
print(f"  Grand Total:                ${recession_exp + non_recession_exp:>12,.0f}")
print("-" * 55)
print(f"  Recession Share:            {recession_exp/(recession_exp+non_recession_exp)*100:.1f}%")
print(f"  Non-Recession Share:        {non_recession_exp/(recession_exp+non_recession_exp)*100:.1f}%")
print("=" * 55)
````

### TASK 1.8: Develop a pie chart to display the total Advertisement expenditure for each vehicle type during recession period.<br>

````python
# Task 1.8: Pie Chart - Ad Expenditure by Vehicle Type During Recession

# Filter for recession periods only
df_recession = df[df['Recession'] == 1]

# Calculate total advertising expenditure per vehicle type
vehicle_exp = df_recession.groupby('Vehicle_Type')['Advertising_Expenditure'].sum()

# Sort values for better visualization
vehicle_exp = vehicle_exp.sort_values(ascending=False)

# ── Colors and Explode ────────────────────────────────────────────────────────
colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', 
          '#FFEAA7', '#DDA0DD', '#98D8C8', '#F7DC6F']

# Explode the largest slice for emphasis
explode = [0.1 if i == 0 else 0 for i in range(len(vehicle_exp))]

# ── Figure with Two Charts ────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(18, 8))

# ── Pie Chart 1: Standard Pie ─────────────────────────────────────────────────
wedges, texts, autotexts = axes[0].pie(
    vehicle_exp.values,
    labels=vehicle_exp.index,
    colors=colors[:len(vehicle_exp)],
    explode=explode,
    autopct='%1.1f%%',
    shadow=True,
    startangle=140,
    textprops={'fontsize': 11},
    wedgeprops={'edgecolor': 'white', 'linewidth': 2},
    pctdistance=0.80
)

# Style percentage text
for autotext in autotexts:
    autotext.set_fontsize(11)
    autotext.set_fontweight('bold')
    autotext.set_color('white')

# Style label text
for text in texts:
    text.set_fontsize(11)
    text.set_fontweight('bold')

axes[0].set_title('Advertisement Expenditure\nby Vehicle Type During Recession',
                  fontsize=13, fontweight='bold', pad=20)

# ── Pie Chart 2: Donut Chart ──────────────────────────────────────────────────
wedges2, texts2, autotexts2 = axes[1].pie(
    vehicle_exp.values,
    labels=None,                            # labels in legend instead
    colors=colors[:len(vehicle_exp)],
    autopct='%1.1f%%',
    startangle=140,
    textprops={'fontsize': 10},
    wedgeprops={'edgecolor': 'white', 
                'linewidth': 2,
                'width': 0.6},             # makes it a donut chart
    pctdistance=0.75
)

# Style percentage text
for autotext in autotexts2:
    autotext.set_fontsize(10)
    autotext.set_fontweight('bold')
    autotext.set_color('white')

# Add center text for donut chart
total = vehicle_exp.sum()
axes[1].text(0, 0,
             f'Total\n${total:,.0f}',
             ha='center',
             va='center',
             fontsize=12,
             fontweight='bold',
             color='#2C3E50')

axes[1].set_title('Advertisement Expenditure Share\nby Vehicle Type (Donut View)',
                  fontsize=13, fontweight='bold', pad=20)

# Add legend for donut chart
axes[1].legend(
    wedges2,
    [f'{vtype}: ${val:,.0f}' for vtype, val in zip(vehicle_exp.index, vehicle_exp.values)],
    title='Vehicle Type: Total Expenditure',
    title_fontsize=11,
    loc='lower center',
    bbox_to_anchor=(0.5, -0.18),
    ncol=2,
    fontsize=9,
    framealpha=0.8
)

# ── Overall Figure Title ──────────────────────────────────────────────────────
fig.suptitle('Total Advertisement Expenditure by Vehicle Type\nDuring Recession Period',
             fontsize=15, fontweight='bold', y=1.02)

plt.tight_layout()
plt.show()

# ── Print Summary Table ───────────────────────────────────────────────────────
print("=" * 60)
print("  Ad Expenditure by Vehicle Type During Recession")
print("=" * 60)
print(f"  {'Vehicle Type':<25} {'Expenditure':>12}  {'Share':>8}")
print("-" * 60)
for vtype, val in vehicle_exp.items():
    share = (val / total) * 100
    print(f"  {vtype:<25} ${val:>11,.0f}  {share:>7.1f}%")
print("-" * 60)
print(f"  {'Grand Total':<25} ${total:>11,.0f}  {'100.0%':>8}")
print("=" * 60)
````

````python
#import pandas as pd
#import matplotlib.pyplot as plt

# Load data
url = ("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/"
       "IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/"
       "historical_automobile_sales.csv")

df = pd.read_csv(url)

# Filter Recession periods only
recession_df = df[df["Recession"] == 1]

# Aggregate mean sales grouped by unemployment_rate and Vehicle_Type
grouped = (recession_df
           .groupby(["unemployment_rate", "Vehicle_Type"])["Automobile_Sales"]
           .mean()
           .reset_index())

# Plot
fig, ax = plt.subplots(figsize=(12, 6))

for vtype in grouped["Vehicle_Type"].unique():
    subset = grouped[grouped["Vehicle_Type"] == vtype].sort_values("unemployment_rate")
    ax.plot(subset["unemployment_rate"], subset["Automobile_Sales"],
            marker="o", linewidth=2, label=vtype)

ax.set_title("Effect of Unemployment Rate on Vehicle Type and Sales",
             fontsize=14, fontweight="bold")
ax.set_xlabel("Unemployment Rate (%)", fontsize=12)
ax.set_ylabel("Average Automobile Sales", fontsize=12)
ax.legend(title="Vehicle Type", bbox_to_anchor=(1.01, 1), loc="upper left")
ax.grid(axis="y", linestyle="--", alpha=0.5)

plt.tight_layout()
plt.show()
````
